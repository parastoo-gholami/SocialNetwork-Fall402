```python
def get_video_stats(video_id):
    video_info = youtube.videos().list(part='statistics', id=video_id).execute()
    if 'items' in video_info and video_info['items']:
        statistics = video_info['items'][0]['statistics']
        comment_count = int(statistics['commentCount']) if 'commentCount' in statistics else 0
        view_count = int(statistics['viewCount']) if 'viewCount' in statistics else 0
        likes = int(statistics['likeCount']) if 'likeCount' in statistics else 0
        dislikes = int(statistics['dislikeCount']) if 'dislikeCount' in statistics else 0
        return comment_count, view_count, likes, dislikes
    else:
        return 0, 0, 0, 0

df_video[['Comment Count', 'View Count', 'Likes', 'Dislikes']] = df_video['Video ID'].apply(lambda x: pd.Series(get_video_stats(x)))
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download the 'punkt' and 'stopwords' packages
nltk.download('punkt')
nltk.download('stopwords')

# Your existing code
comments_text = df_comments['Comment Text'].str.lower().str.replace(r'\b\w\b', '').apply(word_tokenize)

stop_words = set(stopwords.words('english'))
filtered_comments_text = comments_text.apply(lambda x: [word for word in x if word.isalpha() and word not in stop_words])

all_words = [word for sublist in filtered_comments_text for word in sublist]
fdist = FreqDist(all_words)

top_keywords = fdist.most_common(25)
```
